```python
#look into the distribution of the tracks dataset before doing more
zero_to_ten = 0
ten_to_twenty = 0
twenty_to_thirty = 0
thirty_to_fourty = 0
fourty_to_fifty = 0
fifty_to_sixty = 0
sixty_to_seventy = 0
seventy_to_eighty = 0
eighty_to_ninety = 0
ninety_to_hundred = 0

for item in kanye_tracks['popularity']:
  if item >= 0 and item <= 10:
    zero_to_ten = zero_to_ten + 1
  elif item > 10 and item <=20:
    ten_to_twenty = ten_to_twenty + 1
  elif item > 20 and item <= 30:
    twenty_to_thirty = twenty_to_thirty + 1
  elif item > 30 and item <=40:
    thirty_to_fourty = thirty_to_fourty + 1
  elif item > 40 and item <=50:
    fourty_to_fifty = fourty_to_fifty + 1
  elif item > 50 and item <=60:
    fifty_to_sixty = fifty_to_sixty + 1
  elif item > 60 and item <=70:
    sixty_to_seventy = sixty_to_seventy + 1
  elif item > 70 and item <=80:
    seventy_to_eighty = seventy_to_eighty + 1
  elif item > 80 and item <=90:
    eighty_to_ninety = eighty_to_ninety + 1
  else:
    ninety_to_hundred = ninety_to_hundred + 1
```

```python
print("0-10: ", zero_to_ten)
print("10-20: ", ten_to_twenty)
print("20-30: ", twenty_to_thirty)
print("30-40: ", thirty_to_fourty)
print("40-50: ", fourty_to_fifty)
print("50-60: ", fifty_to_sixty)
print("60-70: ", sixty_to_seventy)
print("70-80: ", seventy_to_eighty)
print("80-90: ", eighty_to_ninety)
print("90-100: ", ninety_to_hundred)
```
